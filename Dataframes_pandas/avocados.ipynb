```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
avocados = pd.read_csv('avocados.csv')
# Look at the first few rows of data
avocados.head()
```

```python
avocados.columns
```

```python
# Get the total number of avocados sold of each size
nb_sold_by_size = avocados.groupby(avocados['size'])[['nb_sold']].sum()
```

```python
# Create a bar plot of the number of avocados sold by size
nb_sold_by_size.plot(kind='bar')
# Show the plot
plt.show()
```

```python
# Get the total number of avocados sold on each date
nb_sold_by_date = avocados.groupby('date')['nb_sold'].sum()

# Create a line plot of the number of avocados sold by date
nb_sold_by_date.plot(kind='line')

# Show the plot
plt.show()
```

```python
# Scatter plot of avg_price vs. nb_sold with title
avocados.plot(x= "nb_sold", y="avg_price", kind="scatter", title = "Number of avocados sold vs. average price")

# Show the plot
plt.show()
```

```python
# Modify bins to 20
avocados[avocados["type"] == "conventional"]["avg_price"].hist(alpha=0.5, bins=20)

# Modify bins to 20
avocados[avocados["type"] == "organic"]["avg_price"].hist(alpha=0.5, bins=20)

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()
```

First off, all the values in x-axis are scaled by factor of 1e7 (10^7).<br>
Some cases it is 1e6 (10^6)<br>
Notice, in the above histograms, there are no data points for 0. They all start at different points. For ex, <br>
fig1 starts at 0.75*10^7<br>
fig2 starts at 0.75*10^7<br>
fig3 starts at 0.5*10^7<br>

```python
# Fill in missing values with 0
avocados_filled = avocados_2016.fillna(0)

# Create histograms of the filled columns
avocados_filled[cols_with_missing].hist()

# Show the plot
plt.show()
```
