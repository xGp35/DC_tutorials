```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(123)
```

```python
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)
```

```python
print(random_walk)
```

```python
plt.plot(random_walk)
plt.show()
```

```python
# initialize and populate all_walks
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() < 0.001:
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)
```
